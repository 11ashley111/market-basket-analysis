```python
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)

df['InvoiceNo']=df['InvoiceNo'].astype('str')
df=df[~df['InvoiceNo'].str.contains('C')]
df.shape
```

```python
df.InvoiceNo.nunique()
```

```python
top_10=df.groupby('Country').nunique().sort_values('InvoiceNo',ascending =False).head(10)
```

```python
top_10
```

```python
data= (df[df['Country']=='United Kingdom'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0)
       .set_index('InvoiceNo'))
data.head(2)       
```

```python
data.shape
```

```python
def encode_units(x):
    return 1 if x > 0 else 0

data = data.apply(lambda x: x.map(encode_units))
```

```python
data
```

```python
data=data[(data>0).sum(axis=1) >= 2]
data
```

```python

import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Generate frequent itemsets
frequent_itemsets_plus = apriori(data, min_support=0.03, use_colnames=True)

# Sort by support
frequent_itemsets_plus = frequent_itemsets_plus.sort_values('support', ascending=False).reset_index(drop=True)

# Add length column
frequent_itemsets_plus['length'] = frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))

# Display the DataFrame
print(frequent_itemsets_plus)
```

```python
as_rules = association_rules(frequent_itemsets_plus, metric='lift',
                              min_threshold=1).sort_values('lift',  ascending=False).reset_index(drop=True)
as_rules.head(10)
```
